```python
tobe_sent = """
const memosList = document.getElementById("memos-list");
const btnAdd = document.getElementById("add-btn");
const fieldTitle = document.getElementById("title");
const fieldContent = document.getElementById("text");

var memos = [];
var inc = 0;

btnAdd.onclick = add;

function add(event) {
  memos.push({
    title: fieldTitle.value,
    content: fieldContent.value,
    date: new Date(Date.now()),
    id: inc,
  });

  const memoItemDiv = document.createElement("div");
  memoItemDiv.className = "memo-item";
  memoItemDiv.id = `${inc}_item`;

  const memoHeaderDiv = document.createElement("div");
  memoHeaderDiv.className = "memo-header";

  memoItemDiv.appendChild(memoHeaderDiv);

  const titleH2 = document.createElement("h2");
  const dateP = document.createElement("p");
  const btnDelete = document.createElement("button");
  btnDelete.innerText = "Delete";
  btnDelete.id = `${inc}`;
  btnDelete.onclick = deleteItem;

  memoItemDiv.appendChild(titleH2);
  memoHeaderDiv.appendChild(dateP);
  memoHeaderDiv.appendChild(btnDelete);

  const contentP = document.createElement("p");

  dateP.innerHTML = new Date(Date.now());
  titleH2.innerText = fieldTitle.value;
  contentP.innerText = fieldContent.value;

  memoItemDiv.appendChild(contentP);

  memosList.appendChild(memoItemDiv);

  localStorage.setItem("data", JSON.stringify(memos));

  inc++;
}

function deleteItem(event) {
  memos = memos.filter((val, idx, arr) =>
    val.id != event.target.id ? val : null
  );

  console.log(memos);
  localStorage.setItem("data", JSON.stringify(memos));

  const toBeRemoved = document.getElementById(`${event.target.id}_item`);
  memosList.removeChild(toBeRemoved);
}

function retriveAndRender() {
  let data = JSON.parse(localStorage.getItem("data"));

  memos = data == null ? memos : data;

  data.forEach((element) => {
    const memoItemDiv = document.createElement("div");
    memoItemDiv.className = "memo-item";
    memoItemDiv.id = `${element.id}_item`;

    const memoHeaderDiv = document.createElement("div");
    memoHeaderDiv.className = "memo-header";

    memoItemDiv.appendChild(memoHeaderDiv);

    const titleH2 = document.createElement("h2");
    const dateP = document.createElement("p");
    const btnDelete = document.createElement("button");
    btnDelete.innerText = "Delete";
    btnDelete.id = `${element.id}`;
    btnDelete.onclick = deleteItem;

    memoItemDiv.appendChild(titleH2);
    memoHeaderDiv.appendChild(dateP);
    memoHeaderDiv.appendChild(btnDelete);

    const contentP = document.createElement("p");

    dateP.innerHTML = element.date ? element.date : "No date saved";

    titleH2.innerText = element.title;
    contentP.innerText = element.content;

    memoItemDiv.appendChild(contentP);

    memosList.appendChild(memoItemDiv);

    inc = element.id;
  });
}

retriveAndRender();

"""
```
